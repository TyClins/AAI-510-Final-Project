```python
import pandas as pd
import numpy as np
```

```python
# Read in data
df = pd.read_csv("data/k2pandc_2024.05.30_19.37.55.csv", comment='#')
df.head()
```

```python
# Check for missing values and data types
df.info()
```

```python
df.disposition.value_counts()
```

```python
confirmed = df[df.disposition.isin(["CONFIRMED", "FALSE POSITIVE"])]
```

```python
# Replace disposition values with planet flag
condition = [True if disp == "CONFIRMED" else False for disp in confirmed.disposition]
confirmed.insert(2, "planet", condition)
data=confirmed.drop(["disposition", "pl_name"], axis=1)
```

```python
data.head()
```

```python
# Get and display the correlations
correlations = data.corr()["planet"].sort_values(ascending = False)
print('Correlations:\n', correlations)
```
